<i>Here I am filtering all the columns for a particular team</i>

```python
Team = input("Input the Team: ")
data[(data["HomeTeam"] == Team) | (data["AwayTeam"] == Team)]
```

<i>Filtering all the game played by 2 teams</i>

```python
HTeam = input("Input the Home Team: ")
ATeam = input("Input the Away Team: ")
data[((data["HomeTeam"] == HTeam) & (data["AwayTeam"] == ATeam)) | ((data["HomeTeam"] == ATeam) & (data["AwayTeam"] == HTeam))]
```

Adding a new row "FHTG - FTAG" to the data

```python
data["FTHG - FTAG"] = data["FTHG"] - data["FTAG"]
data.head()
```

Making sure row "FHTG - FTAG" is only positive

```python
data["FTHG - FTAG"] = data["FTHG - FTAG"].abs()
data[:50]
```

Taking the mean of column "FTHG - FTAG"

```python
mean_fthg_ftag = data["FTHG - FTAG"].mean()
print(f"The mean of the column [FTHG - FTAG] is: {mean_fthg_ftag}")
```
